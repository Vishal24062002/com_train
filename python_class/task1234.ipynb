```python
# Write a Python program to do the following: 
# • Accept lengths L1 and L2 for two lists from the user. 
# • Create List A of length L1 and ensure that there are no repeating elements 
# • Create List B of length L2 and ensure that there are no repeating elements 
# • Perform the following: o Create list C that contains elements common to lists A and B. 
# a. Create list D that contains elements unique to lists A and B. 
# b. Print list C 
# c. Print list D 
# d. Print C sorted in ascending order.
# e. Print D sorted in descending order


# # • Accept lengths L1 and L2 for two lists from the user. 
# • Create List A of length L1 and ensure that there are no repeating elements 
# • Create List B of length L2 and ensure that there are no repeating elements 

l1=int(input("enter the length of list 1"))
def create(n):
    li=[]   
    while len(li)<n:
        ele=int(input("enter the elements"))
        if ele not in li:
          li.append(ele)
        else:
            pass
    return li
listA=create(l1)
print(listA)
l2=int(input("enter  the length of second list"))
listB=create(l2)
print(listB)

# Perform the following: o Create list C that contains elements common to lists A and B.

listC=[i for i in listA if i  in listB ]

# Create list D that contains elements unique to lists A and B. 
d=listA+listB

listD=[i for i in d if  d.count(i)==1 ]
print("comman element")
print(listC)
print("unique element")
print(listD)


# Print C sorted in ascending order
listC.sort()
print(listC)
# Print D sorted in descending order
listD.sort(reverse=True)
print(listD)
```

```python


#  Given two strings S1 and S2, return a new string that contains the first, middle and last characters of both 
# strings. 
# • Input: 
# o America 
# o Japan 
# • Output: 
# o AJrpan 


def combine(s1, s2):
    def get_middle(s):
        mid = len(s) // 2
        return s[mid]
 
    
    firsts1 = s1[0]
    middles1 = get_middle(s1)
    lasts1 = s1[-1]
    
    firsts2 = s2[0]
    middles2 = get_middle(s2)
    lasts2 = s2[-1]
    
    result = firsts1 +  firsts2 + middles1 +  middles2 + lasts1+lasts2
    return result


s1 = "America"
s2 = "Japan"
output = combine(s1, s2)
print(output) 
```

```python
# task 5-->
# Write a Python program to count the number of strings where the string length is 2 or more and the first and 
# last character are same from a given list of strings.
# Example:
# Input ➔ ['abc', 'xyz', 'aba', '1221'] 
# Output ➔ 2

def count(l):
    count = 0
    for i in l:
        if len(i) >= 2 and i[0] == i[-1]:
            count += 1
    return count

l=['abc', 'xyz', 'aba', '1221'] 
cnt=count(l)
print(cnt)
```

```python
#8- Consider the following situation: 
# • Headings = [“first_name”, “last_name”, “age”] 
# • Person1 = [“Mark”,”Taylor”,20] 
# • Person2 = [“Emma”,”Jones”,30] 
# Write a Python program that will create the output dictionary where each element of the headings is the key 
# and each corresponding element in person lists will be a list of values. 
# Output: 
# {'first_name': ['Mark', 'Emma'], 'last_name': ['Taylor', 'Jones'], 'age': [20, 30]}



Headings = ["first_name", "last_name", "age"] 
Person1 = ["Mark","Taylor",20] 
Person2 = ["Emma","Jones",30] 

people=[Person1,Person2]

res={key:[] for key in Headings}
for i in people:
    for key,value in zip(Headings,i):
        res[key].append(value)
print(res)

```
